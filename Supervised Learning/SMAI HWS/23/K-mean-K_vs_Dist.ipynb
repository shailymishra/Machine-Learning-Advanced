```python
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
X=np.array(np.random.random((10,2)))
kmeans = KMeans(n_clusters=2,init='k-means++',max_iter=600,random_state=0,algorithm='auto').fit(X)
plt.scatter(X[:,0],X[:,1],c=kmeans.labels_.astype(float))
print(kmeans.cluster_centers_)
#red denotes the centroid ..illustrating only for 2 centroids   
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],c='r')
plt.title("2-D data")
plt.show()
```

```python
#to evaluate k means
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
X=np.array(np.random.random((10,2)))

sse = {}
for i in range(1,11):
    kmeans = KMeans(n_clusters=i,init='k-means++',max_iter=600,random_state=0,algorithm='auto').fit(X)
    sse[i]=kmeans.inertia_ # .inertia_ gives the sum of squared distances of the samples to their closes
plt.plot(list(sse.keys()), list(sse.values()))
plt.title("iteration vs sum of squared error")
plt.xlabel("# iterations")
plt.ylabel("value of sse")
#clearly we can see the sum of squared error which the Kmean algorithm aims to minimize is max for k=1
```
